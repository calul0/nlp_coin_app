### Hufigkeiten aller Wrter

Die Tabelle stellt fr jedes Wort:
  - total - die Anzahl der gesamt Hufigkeit,
  - hit - die Anzahl der predictions, welche in der GT erfllt sind
  - miss - die Anzahl der predictions, die nicht in der GT erfllt sind
  - failed - die Anzahl der predictions, die nicht in der GT identifiziert werden konnten
  - acc - die Accuracy mit der Formel acc= hit/total, wenn hit,total > 0 und -1 wenn failed > 0, ansonsten 0
  - notInGt: listet alle Design_Ids, die hier in failed betroffen sind
  - missDesignId: listet alle DesignIds, die nicht predicted wurden.

### Hufigkeiten aller Relationen

  - missingRelation - die Anzahl der Relation, die in prediction auftauchte, allerdings in GT nicht identifiziert werden konnte,
